```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Analytical solution
arr = np.arange(0,10,np.pi/6)
dx=arr[1]-arr[0]
x = np.sin(arr)
y = np.cos(arr)
print(x,y)
```

```python
#First order left derivative
left=[]
for i in range(1,len(x)):
    left.append((x[i]-x[i-1])/dx)

print(left)
print(len(left))
```

```python
#First order right derivative
right = []
for i in range(0,len(x)-1):
    right.append((x[i+1]-x[i])/dx)
#right.append(0)    
print(right)
print(len(right))
```

```python
#Central difference 2nd order
central = []
for i in range(1,len(x)-1):
    central.append((x[i+1]-x[i-1])/(2*dx))

print(central)

c2=x[1:]
```

```python
#Central difference 4th order
c4=[]
for i in range(2,len(x)-2):
    c4.append((-x[i+2]+8*x[i+1]-8*x[i-1]+x[i-2])/(12*dx))

print(c4)
```

```python
plt.plot(arr,y,lw=4, color='yellow', label='Analytical')
plt.plot(arr[:-1], left, label='Left Derivative')
plt.plot(arr[1:], right, label='Right Derivative')
plt.plot(arr[1:-1], central, label='Central 2nd Order')
plt.plot(arr[2:-2], c4, marker='+', label='Central 4th Order')
plt.legend()
plt.title("Comparison of First Derivative Approximates, f(x)=sinx")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.show()
```
